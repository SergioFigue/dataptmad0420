#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
y1 = x**2
y2 = np.exp(x)


fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(12,8))

axes[0].plot(x, y1, label='x**2', linestyle='-', color='blue')
axes[0].legend(fontsize=15, shadow=True, fancybox=False)
axes[0].set(xlabel='x',
           ylabel='x**2')
axes[0].set_yscale('linear')
axes[0].set_xscale('linear')
axes[0].set_title(label='Normal scale', fontsize=20)

axes[1].plot(x, y2, label='y=e(x)', linestyle='-', color='#00FFFF')
axes[1].legend(fontsize=15, shadow=True, fancybox=False)
axes[1].set(xlabel='x',
           ylabel='y=e(x)')
axes[1].set_yscale('log')
axes[1].set_xscale('log')
axes[1].set_title(label='Logarithmic scale', fontsize=20)

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
steps = fitbit.groupby('Months_encoded', as_index=False)[['Steps']].agg('mean').astype(int)
Months = steps['Months_encoded'].tolist()
Steps = steps['Steps'].tolist()

steps2 = fitbit.groupby('Work_or_Weekend', as_index=False)['Steps'].mean().astype(int)
Days = steps2['Work_or_Weekend'].tolist()
WSteeps = steps2['Steps'].tolist()
```

```python
import os

# MEDIAN STEPS BY MONTH_ENCODED

fig, ax = plt.subplots(2, 1, figsize=(12,4))

fig.suptitle("""Steps per Month and day""", fontweight ="bold")

plt.sca(ax[0]) 
plt.bar(Months, Steps)
plt.setp(ax[0].get_xticklabels(), rotation=45, horizontalalignment='right')

# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.sca(ax[1])
plt.barh(Days, WSteeps)
plt.setp(ax[1].get_xticklabels(), rotation=45, horizontalalignment='right');

plt.show()  


if not os.path.exists('../figures/'):
    os.mkdir('../figures/')


fig.savefig('../figures/Fitbit_plot.pdf', 
            transparent=False, 
            dpi=80, 
            bbox_inches="tight")
```

#### Plot the correlation matrix using seaborn.

```python
corr = titanic.corr()

mask = np.triu(np.ones_like(corr, dtype=np.bool))

f, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(corr, mask=mask, vmax=.3, center=0,
            square=True, linewidths=.5);


```
